```python
import spectrochempy as scp
import pandas as pd
import numpy as np
import glob
import os
```

```python
folder_path = r'C:\Users\noamkaro\OneDrive - Technion\Nucleation research\FTIR\RefsAu'
final_csv_path = r'C:\Users\noamkaro\OneDrive - Technion\Nucleation research\FTIR\RefsAu\refs_Au.csv'
spa_files = glob.glob(os.path.join(folder_path, '*.SPA'))

dataframes = []
for spa_file_path in spa_files:
    dataset = scp.read(spa_file_path)
    data = dataset.data.T
    df = pd.DataFrame(data, columns=dataset.x_labels)
    wavenumbers = np.linspace(4000.1885, 649.9040, 6950)
    df.insert(0, 'Wavenumber', wavenumbers)
    dataframes.append(df)


final_df = dataframes[0]  # Initialize with the first dataframe

for idx, df in enumerate(dataframes[1:], start=1):
    # Rename the second column of the DataFrame to ensure uniqueness
    new_column_name = os.path.splitext(os.path.basename(spa_files[idx]))[0]
    df_renamed = df.rename(columns={df.columns[1]: new_column_name})
    
    # Merge on 'Wavenumber', now with a uniquely named second column
    final_df = pd.merge(final_df, df_renamed[['Wavenumber', new_column_name]], on='Wavenumber', how='outer')

# After merging, the 'Wavenumber' column will be followed by uniquely named columns based on the file names
final_df.to_csv(final_csv_path, index=False)

```

```python
from colorsys import hls_to_rgb
def get_distinct_colors(n):

    colors = []

    for i in np.arange(0., 360., 360. / n):
        h = i / 360.
        l = (50 + np.random.rand() * 10) / 100.
        s = (90 + np.random.rand() * 10) / 100.
        colors.append(hls_to_rgb(h, l, s))

    return colors

colors = get_distinct_colors(25)
```
